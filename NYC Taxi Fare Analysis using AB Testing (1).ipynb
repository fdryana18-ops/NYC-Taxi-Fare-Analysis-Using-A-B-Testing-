```python
#IMPORT PACKAGES
import pandas as pd
from scipy import stats
#READ DATA
taxi_data= pd.read_csv("C:/Users/USER/Downloads/2017_Yellow_Taxi_Trip_Data.csv", index_col = 0)
# Note that in dataset, the payment method encode as 1: Credit card, 2: Cash, 3: No charge, 4: Dispute, 5: Unknown
#ANALYZE AND CONSTRUCT
# descriptive stats code for EDA
taxi_data.describe(include='all')
#We look at the average fare amount for each payment type since we only interested in payment method and fare amount
taxi_data.groupby('payment_type')['fare_amount'].mean()
#average payment method by credit card is the highest.
#conduct hypothesis testing
#hypothesis test, A/B test
#significance level
credit_card = taxi_data[taxi_data['payment_type'] == 1]['fare_amount']
cash = taxi_data[taxi_data['payment_type'] == 2]['fare_amount']
stats.ttest_ind(a=credit_card, b=cash, equal_var=False)
#we reject h null since p value< test statistic
#there is a statistically significant difference in the average fare amount between customers who use credit cards and customers who use cash.

```
